```python
import numpy as np
import matplotlib.pyplot as plt

# dx Spatial step in ft
# dt Time step in hours
def heatmap(dx, dt):

    # Given parameters
    Ti = 100.0  # Initial temperature in °F
    Ts = 300.0  # Surface temperature in °F
    alpha = 0.1  # Diffusivity in ft^2/hour
    tf = 0.1  # Final time in hours
    if ((alpha*dt)/dx**2) > 0.5:
      exit()
# Calculate the number of spatial and temporal points
    N = int(1 / dx) + 1
    M = int(tf / dt) + 1

    # Initialize temperature grid
    u = np.zeros((N, M))
    u[:, 0] = Ti  # Initial condition

    # Update boundary conditions
    u[0, :] = Ts
    u[-1, :] = Ts

    # Solve the heat equation iteratively
    for n in range(M - 1):
      for i in range(1, N - 1):
        u[i, n + 1] = u[i, n] + alpha * dt / dx ** 2 * (u[i + 1, n] - 2 * u[i, n] + u[i - 1, n])

 # Extract the temperature distribution at the final time
    temperature_final = u[:, -1]

     # Print the approximate temperature distribution at the final time
    print("Approximate Temperature Distribution at t = 0.1 hour:")
    for i, temp in enumerate(temperature_final):
      print("Position x = {:.2f} ft: Temperature = {:.2f} °F".format(i * dx, temp))

# Plot the temperature distribution
    x = np.linspace(0, 1, N)
    plt.plot(x, temperature_final)
    plt.xlabel("Position (ft)")
    plt.ylabel("Temperature (°F)")
    plt.title("Temperature Distribution at t = 0.1 hour")
    plt.grid(True)
    plt.show()

print("Letra A")
heatmap(0.05, 0.01)
heatmap(0.05,0.05)
```

```python
import numpy as np
import sys
import matplotlib.pyplot as plt

def heatmap(DX, DT):
  TIME_MAX = 0.1
  ALPHA = 0.1

  numSteps = int(TIME_MAX / DT)
  numPoints = int(1 / DX) + 1

  T = np.zeros((numSteps + 1, numPoints))
  if ((ALPHA*DT)/DX**2) > 0.5:
    print("erro")
    sys.exit()

  # Initial conditions
  T[0, :] = 100.0

  # Boundary conditions
  T[:, 0] = 300.0
  T[:, -1] = 300.0

  # Calculation of temperature distribution
  for t in range(numSteps):
      for i in range(1, numPoints - 1):
        r = ALPHA * DT / (DX**2)
        T[t + 1, i] = T[t, i] + r * (T[t, i + 1] - 2 * T[t, i] + T[t, i - 1])

                      # Printing the temperature distribution at final time
  x = np.linspace(0, 1, numPoints)

  # Plotting the temperature distribution
  plt.figure()
  plt.plot(x, T[-1, :], label=f't = {TIME_MAX} hours')
  plt.xlabel('x')
  plt.ylabel('Temperature')
  plt.title('Temperature Distribution')
  plt.grid(True)
  plt.legend()
  plt.show()

heatmap(0.05, 0.01)
heatmap(0.05, 0.05)
```
